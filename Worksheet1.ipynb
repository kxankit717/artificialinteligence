```python
# Define size
rows, cols = 5, 5

# Create a zeros array
arr = np.zeros((rows, cols), dtype=int)

# Set borders to 1
arr[0, :] = 1       # Top row
arr[-1, :] = 1      # Bottom row
arr[:, 0] = 1       # Left column
arr[:, -1] = 1      # Right column

print(arr)
```

```python

checkerboard = np.zeros((8, 8), dtype=int)

checkerboard[1::2, ::2] = 1  # Rows 1,3,5,7, columns 0,2,4,6
checkerboard[::2, 1::2] = 1  # Rows 0,2,4,6, columns 1,3,5,7

print(checkerboard)
```

```python
#problem3
import numpy as np

# Arrays
x = np.array([[1, 2], [3, 5]])
y = np.array([[5, 6], [7, 8]])
v = np.array([9, 10])
w = np.array([11, 12])

# Add
add_xy = x + y
print("Addition of x and y:\n", add_xy)

# Subtract
sub_xy = x - y
print("Subtraction of x and y:\n", sub_xy)

# Multiply x by 3
mul_x = x * 3
print("x multiplied by 3:\n", mul_x)

# Square of each element in x
square_x = x ** 2
print("Square of elements in x:\n", square_x)

# Dot product v and w
dot_vw = np.dot(v, w)
print("Dot product of v and w:", dot_vw)

# Dot product x and v
dot_xv = np.dot(x, v)
print("Dot product of x and v:\n", dot_xv)

# Dot product x and y
dot_xy = np.dot(x, y)
print("Dot product of x and y:\n", dot_xy)

# Concatenate x and y along rows (axis=0)
concat_xy_row = np.concatenate((x, y), axis=0)
print("Concatenate x and y along rows:\n", concat_xy_row)

# Concatenate v and w along columns (make them 2D first)
v_2d = v.reshape(1, 2)
w_2d = w.reshape(1, 2)
concat_vw_col = np.concatenate((v_2d.T, w_2d.T), axis=1)
print("Concatenate v and w along columns:\n", concat_vw_col)




```

```python
#Problem 4 (Matrix Operation)

# Arrays
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

# Compute inverse of A
A_inv = np.linalg.inv(A)

# Multiply A and A_inv
result = np.dot(A, A_inv)

print("A × A⁻¹:\n", result)

# Identity matrix for comparison
I = np.eye(2)
print("Identity matrix:\n", I)

# Check if equal
print("Is A × A⁻¹ is identity matrix", np.allclose(result, I))
AB = np.dot(A, B)
BA = np.dot(B, A)

print("A × B:\n", AB)
print("B × A:\n", BA)
print("Is A × B equal to B × A?", np.array_equal(AB, BA))


AB_T = (np.dot(A, B)).T
B_TA_T = np.dot(B.T, A.T)

print("(AB)ᵀ:\n", AB_T)
print("Bᵀ × Aᵀ:\n", B_TA_T)
print("Is (AB)ᵀ equal to Bᵀ × Aᵀ?", np.array_equal(AB_T, B_TA_T))

# Coefficient matrix A
A_eq = np.array([[2, -3, 1],
                 [1, -1, 2],
                 [3, 1, -1]])

# Constants vector B
B_eq = np.array([-1, -3, 9])

# Compute inverse of A_eq
A_inv_eq = np.linalg.inv(A_eq)

# Solve for X
X = np.dot(A_inv_eq, B_eq)

print("Solution [x, y, z]:", X)


```
